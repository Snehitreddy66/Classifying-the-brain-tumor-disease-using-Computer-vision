```python
SIZE = 224
```

```python
train_images = []
train_labels = [] 
for directory_path in glob.glob("C:\\Users\\SNEHIT REDDY\\Desktop\\Brain Tumor\\Brain tumor Dataset\\train\\*"):
    label = directory_path.split("\\")[-1]
    print(label)
    for img_path in glob.glob(os.path.join(directory_path, "*.png")):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)       
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        train_images.append(img)
        train_labels.append(label)
        
train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
# test
test_images = []
test_labels = [] 
for directory_path in glob.glob("C:\\Users\\SNEHIT REDDY\\Desktop\\Brain Tumor\\Brain tumor Dataset\\val\\*"):
    fruit_label = directory_path.split("\\")[-1]
    for img_path in glob.glob(os.path.join(directory_path, "*.png")):
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        test_images.append(img)
        test_labels.append(fruit_label)
        
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
from keras import optimizers
sgd = optimizers.SGD(learning_rate=0.1, decay=1e-6, momentum=0.9)
print("[INFO] training network...")
model.compile(loss="categorical_crossentropy", optimizer=sgd,metrics=["accuracy"])
```

```python
custom_early_stopping = EarlyStopping(
    monitor='val_loss', 
    patience=5, 
    min_delta=0.1, 
    mode='min'
)
```

```python
import datetime
start = datetime.datetime.now()
#Fit the model. Do not forget to use on-hot-encoded Y values. 
history=model.fit(x_train, y_train_one_hot, batch_size=128,epochs=10, validation_data=(x_test, y_test_one_hot),callbacks=[custom_early_stopping])

end = datetime.datetime.now()
print("Total execution time : ", end-start)
```

```python
print("[INFO] Calculating model accuracy")
scores = model.evaluate(x_train, y_train_one_hot)
print("Train Loss = " +str(scores[0]))
print("Train Accuracy = " + str(scores[1]*100))
```
